## 1. Values have types

Python values are not all the same. A string that looks like a number is still text until you convert it.

```python
name = "Ava"
score_text = "95.5"
score_number = 95.5

print(type(name))
print(type(score_text))
print(type(score_number))
```

Try this before running: which line will fail, and why?

```python
print(score_number + 10)
# print(score_text + 10)
```

## 2. Clean strings one step at a time

Many data tasks begin with strings that have extra spaces, mixed case, or symbols.

```python
raw_text = "  Open  "

clean_text = raw_text.strip()

print(raw_text)
print(clean_text)
```

```python
raw_category = "  HOME  "

category = raw_category.strip().lower()

print(category)
```

```python
raw_amount = " $1,250.00 "

step_1 = raw_amount.strip()
step_2 = step_1.replace("$", "")
step_3 = step_2.replace(",", "")

print(step_1)
print(step_2)
print(step_3)
```

## 3. Convert text to numbers carefully

`float()` converts numeric text to a number. It fails when the text is blank or not numeric.

```python
amount_text = "1250.00"
amount = float(amount_text)

print(amount)
print(type(amount))
```

```python
examples = ["1250.00", "", "N/A", "abc"]

for value in examples:
    print(value)
```

The next cell introduces `try` and `except`. Use them only around the risky line.

```python
for value in examples:
    try:
        number = float(value)
        print(value, "became", number)
    except ValueError:
        print(value, "is not numeric")
```

## 4. `None` means no usable value

`None` is different from zero. Zero is a real number. `None` means the value is missing, invalid, or not usable for the next step.

```python
valid_zero = 0
missing_value = None

print(valid_zero is None)
print(missing_value is None)
```

Use `is None` when you are checking for `None`.

## 5. Functions package one idea

Start with the smallest possible function: no inputs, one action.

```python
def say_hello():
    print("hello")

say_hello()
```

Now pass an argument into the function. The argument lets the same function work with different values.

```python
def greet(person_name):
    print("hello", person_name)

greet("Ava")
greet("Marcus")
```

A function can return a value. Returned values can be stored, printed, tested, or used in another calculation.

```python
def add_tax(price):
    return price * 1.06

final_price = add_tax(100)
print(final_price)
```

A useful helper function often handles invalid input and returns `None` when it cannot produce a safe answer.

```python
def text_to_number(raw_value):
    if raw_value is None:
        return None

    cleaned = raw_value.strip()
    if cleaned == "" or cleaned.upper() == "N/A":
        return None

    try:
        return float(cleaned)
    except ValueError:
        return None

print(text_to_number("12.5"))
print(text_to_number("N/A"))
```

Small test calls are part of learning. Test the helper before using it inside a larger loop.

```python
test_values = ["10", " 8.5 ", "", "N/A", "oops", None]

for value in test_values:
    print(value, "->", text_to_number(value))
```

## 6. Conditions make decisions

A condition asks a yes or no question. `if`, `elif`, and `else` choose what happens next.

```python
score = 82

if score >= 90:
    print("excellent")
elif score >= 70:
    print("passing")
else:
    print("needs practice")
```

You can combine conditions with `and` when all rules must be true.

```python
has_id = True
amount_is_valid = True

if has_id and amount_is_valid:
    print("keep this row")
else:
    print("reject this row")
```

Use `or` when one bad condition is enough to reject a value.

```python
row_id = ""
amount = None

if row_id == "" or amount is None:
    print("reject")
else:
    print("keep")
```

## 7. Lists store ordered items

A list is useful when you have many values and want to process them in order.

```python
prices = [10.0, 25.5, 4.75]

print(prices[0])
print(len(prices))
```

A loop repeats the same logic for every item in a list.

```python
for price in prices:
    print(price)
```

An accumulator starts with an empty or zero value, then grows inside the loop.

```python
total = 0

for price in prices:
    total = total + price

print(total)
```

Python also has `sum()`, but learners should understand the accumulator pattern first.

```python
print(sum(prices))
```

## 8. Dictionaries store named fields

A dictionary is useful when one thing has several named attributes.

```python
item = {
    "id": "A1",
    "category": "book",
    "price": 12.5,
}

print(item["id"])
print(item["price"])
```

Use `.get()` when a key might be missing.

```python
print(item.get("category"))
print(item.get("state"))
print(item.get("state", "unknown"))
```

A common data shape is a list of dictionaries: many rows, each with named fields.

```python
rows = [
    {"id": "A1", "category": "book", "price": 12.5},
    {"id": "A2", "category": "game", "price": 30.0},
    {"id": "A3", "category": "book", "price": 8.0},
]

print(rows[0])
print(rows[0]["category"])
```

```python
for row in rows:
    print(row["id"], row["category"], row["price"])
```

## 9. Build a clean list from raw rows

This pattern appears in many data engineering tasks:

1. Start with an empty list.
2. Loop over raw rows.
3. Clean and validate one row.
4. Append the clean row only if it passes.

```python
raw_rows = [
    {"id": "A1", "category": " Book ", "price": "12.50"},
    {"id": "A2", "category": "Game", "price": "N/A"},
    {"id": "A3", "category": "book", "price": "8.00"},
]

clean_rows = []

for row in raw_rows:
    item_id = row["id"].strip()
    category = row["category"].strip().lower()
    price = text_to_number(row["price"])

    if item_id == "" or price is None:
        continue

    clean_rows.append({"id": item_id, "category": category, "price": price})

print(clean_rows)
```

`continue` skips the rest of the current loop iteration and moves to the next row.

## 10. Count rejected rows

A rejected row count is another accumulator. It starts at zero and increases when a row fails validation.

```python
rejected_count = 0

for row in raw_rows:
    price = text_to_number(row["price"])

    if price is None:
        rejected_count = rejected_count + 1

print(rejected_count)
```

During development, print the reason. Later, you can remove or comment out the debug print.

```python
for row in raw_rows:
    price = text_to_number(row["price"])

    if price is None:
        print("rejecting", row, "because price is invalid")
```

## 11. Sets track values you have already seen

A set is useful for duplicate detection because membership checks are simple.

```python
seen_ids = set()

print("A1" in seen_ids)
seen_ids.add("A1")
print("A1" in seen_ids)
```

```python
ids = ["A1", "A2", "A1", "A3"]
seen_ids = set()

for item_id in ids:
    if item_id in seen_ids:
        print(item_id, "is a duplicate")
    else:
        seen_ids.add(item_id)
        print(item_id, "is new")
```

## 12. Aggregate by category

Aggregation means grouping rows and calculating totals for each group.

```python
totals_by_category = {}

for row in clean_rows:
    category = row["category"]

    if category not in totals_by_category:
        totals_by_category[category] = 0

    totals_by_category[category] = totals_by_category[category] + row["price"]

print(totals_by_category)
```

Sometimes each group needs several fields, such as count and total.

```python
summary = {}

for row in clean_rows:
    category = row["category"]

    if category not in summary:
        summary[category] = {"count": 0, "total": 0.0}

    summary[category]["count"] = summary[category]["count"] + 1
    summary[category]["total"] = summary[category]["total"] + row["price"]

print(summary)
```

`setdefault` is a shorter version of the same bucket pattern. Learn the longer version first, then this one.

```python
summary = {}

for row in clean_rows:
    bucket = summary.setdefault(row["category"], {"count": 0, "total": 0.0})
    bucket["count"] = bucket["count"] + 1
    bucket["total"] = bucket["total"] + row["price"]

print(summary)
```

## 13. Write a small helper for percentages

Division can fail or mislead when the denominator is zero. Put the guard in one helper function.

```python
def percentage(part, whole):
    if whole == 0:
        return 0.0
    return round(part / whole * 100, 2)

print(percentage(25, 100))
print(percentage(25, 0))
```

## 14. Nested data needs nested loops

Nested data means one value contains another collection. For example, one customer can have many orders.

```python
orders_by_customer = {
    "C1": [10.0, 12.5],
    "C2": [7.0],
    "C3": [20.0, 5.0, 1.5],
}

for customer_id, orders in orders_by_customer.items():
    print(customer_id, orders)
```

```python
for customer_id, orders in orders_by_customer.items():
    customer_total = 0

    for order_amount in orders:
        customer_total = customer_total + order_amount

    print(customer_id, customer_total)
```

Once students understand the nested loop, show the shorter `sum()` version.

```python
for customer_id, orders in orders_by_customer.items():
    print(customer_id, sum(orders))
```

Nested dictionaries are common too. Read the structure from the outside inward.

```python
events_by_customer = {
    "C1": [
        {"kind": "online", "amount": 10.0},
        {"kind": "store", "amount": 12.5},
    ],
    "C2": [
        {"kind": "online", "amount": 7.0},
    ],
}

print(events_by_customer["C1"])
print(events_by_customer["C1"][0])
print(events_by_customer["C1"][0]["amount"])
```

```python
for customer_id, events in events_by_customer.items():
    total = 0

    for event in events:
        total = total + event["amount"]

    print(customer_id, total)
```

## 15. Write files with `pathlib`, `csv`, and `json`

A data task often ends by writing a clean result. This section uses generic demo files.

```python
from pathlib import Path

output_dir = Path("demo_outputs")
output_dir.mkdir(exist_ok=True)

print(output_dir.exists())
```

```python
import csv

csv_path = output_dir / "category_summary.csv"

with open(csv_path, "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["category", "count", "total"])
    for category, bucket in sorted(summary.items()):
        writer.writerow([category, bucket["count"], round(bucket["total"], 2)])

print(csv_path.read_text())
```

```python
import json

json_path = output_dir / "summary.json"

with open(json_path, "w") as file:
    json.dump(summary, file, indent=2)

print(json_path.read_text())
```

## 16. Format output after the logic works

Pretty output is useful, but correctness comes first.

```python
category = "book"
count = 2
total = 20.5

print(f"{category:<12}{count:>6}{total:>12,.2f}")
```

Formatting symbols used above:

- `<12` means left-align in 12 spaces.
- `>6` means right-align in 6 spaces.
- `,.2f` means comma separators and two decimal places.

## 17. Tiny synthesis: a generic clean and summarize flow

This final section combines earlier ideas on a small generic dataset. It is not a template to paste. It is a checkpoint: students should now recognize every line.

```python
raw_orders = [
    {"id": "O1", "category": " Book ", "amount": "10.00"},
    {"id": "O2", "category": "Game", "amount": "N/A"},
    {"id": "O1", "category": "book", "amount": "12.50"},
    {"id": "O3", "category": "Game", "amount": "30.00"},
]

clean_orders = []
seen_order_ids = set()
rejected = 0

for row in raw_orders:
    order_id = row["id"].strip()
    amount = text_to_number(row["amount"])

    if order_id == "" or amount is None or order_id in seen_order_ids:
        rejected = rejected + 1
        continue

    seen_order_ids.add(order_id)
    clean_orders.append({
        "id": order_id,
        "category": row["category"].strip().lower(),
        "amount": amount,
    })

print(clean_orders)
print("rejected", rejected)
```

```python
order_summary = {}

for order in clean_orders:
    bucket = order_summary.setdefault(order["category"], {"count": 0, "total": 0.0})
    bucket["count"] = bucket["count"] + 1
    bucket["total"] = bucket["total"] + order["amount"]

print(order_summary)
```
