## Example 1: Make a filtered copy when it becomes separate work

If you filter rows and plan to work on that filtered table separately, use `.copy()` so your intent is explicit.

```python
example = pd.DataFrame({
    "claim_id": ["A1", "A2", "A3"],
    "status": ["open", "paid", "open"],
    "amount": [100, 500, 900],
})

open_work = example[example["status"] == "open"].copy()
open_work["queue"] = "open_claims_work"
open_work
```

## Example 2: Use `.loc` to update selected rows in the original DataFrame

If you want the original DataFrame to change, build a row mask and assign with `.loc`.

```python
high_amount_mask = example["amount"] >= 500
example.loc[high_amount_mask, "priority"] = "high"
example.loc[~high_amount_mask, "priority"] = "standard"
example
```
