# Activity 5: Groupby Time Series

Analyze cryptocurrency price data by date and coin. This is groupby practice with a time-series flavor.

Solutions notebook for instructor reference. Keep the student drill notebook unchanged.

## Instructor Teaching Note

Emphasize that dates are not just text. Once `data_date` is parsed as a datetime, students can sort by time, create months, and summarize time-based patterns.

```python
import pandas as pd

DATA_PATH = "../Activity_5_Groupby_Time_Series/Resources/crypto_data.csv"
crypto_df = pd.read_csv(DATA_PATH, parse_dates=["data_date"])
crypto_df.head()
```

## 1. Inspect and clean

Drop the columns `data_time` and `timestamp`, then remove rows with missing `data_priceUsd`. Convert `data_priceUsd` to numeric if needed.

```python
clean_crypto_df = crypto_df.drop(columns=["data_time", "timestamp"])
clean_crypto_df = clean_crypto_df.dropna(subset=["data_priceUsd"])
clean_crypto_df["data_priceUsd"] = pd.to_numeric(clean_crypto_df["data_priceUsd"])

print(clean_crypto_df.dtypes)
clean_crypto_df.isna().sum()
```

## 2. Set a datetime index

Set `data_date` as the index and sort by the index.

```python
crypto_ts = clean_crypto_df.set_index("data_date").sort_index()
crypto_ts.head()
```

## 3. Group by coin

Calculate average, minimum, and maximum price for each cryptocurrency.

```python
price_summary = clean_crypto_df.groupby("cryptocurrency")["data_priceUsd"].agg(["mean", "min", "max"])
price_summary
```

## 4. Find volatility by range

Add a `price_range` column to the summary: max price minus min price. Sort from largest range to smallest.

```python
price_summary["price_range"] = price_summary["max"] - price_summary["min"]
price_summary_sorted = price_summary.sort_values("price_range", ascending=False)
price_summary_sorted
```

## 5. Monthly average prices

Create a monthly average price table by cryptocurrency. Keep it simple: group by cryptocurrency and calendar month.

```python
clean_crypto_df["month"] = clean_crypto_df["data_date"].dt.to_period("M")
monthly_prices = clean_crypto_df.groupby(["cryptocurrency", "month"])["data_priceUsd"].mean()
monthly_prices.head(10)
```
