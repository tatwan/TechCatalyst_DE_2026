```python
import pandas as pd

transactions_path = "../Activity_11_Pivot_Crosstab_Melt/Resources/claims_transactions.csv"
monthly_path = "../Activity_11_Pivot_Crosstab_Melt/Resources/monthly_claim_counts_wide.csv"
```

## Example 1: `pivot_table`

Use a pivot table when you want values summarized by two dimensions, such as region by claim type.

```python
example = pd.DataFrame({
    "region": ["East", "East", "West", "West"],
    "claim_type": ["auto", "home", "auto", "home"],
    "amount": [100, 300, 200, 400],
})

pd.pivot_table(example, index="region", columns="claim_type", values="amount", aggfunc="sum", fill_value=0)
```

## Example 2: `melt`

`melt` turns wide columns into a long table. Long format is usually easier to filter, group, and plot.

```python
wide_example = pd.DataFrame({
    "region": ["East", "West"],
    "jan": [10, 8],
    "feb": [12, 11],
})

pd.melt(wide_example, id_vars="region", value_vars=["jan", "feb"], var_name="month", value_name="claims")
```

## Your Turn 1: Load the claim transactions

Read the transactions CSV and inspect rows, shape, and dtypes.

```python
claims_df = pd.read_csv(transactions_path)

claims_df.head()
print(claims_df.shape)
claims_df.dtypes
```

## Your Turn 2: Pivot total amount by region and claim type

Create a pivot table with regions as rows, claim types as columns, and total claim amount as values.

```python
# pivot_table summarizes values into a spreadsheet-like shape.
amount_pivot = claims_df.pivot_table(
    index="region",
    columns="claim_type",
    values="claim_amount",
    aggfunc="sum",
    fill_value=0,
)
amount_pivot
```

## Your Turn 3: Crosstab channel by status

Create a crosstab that counts how many claims appear for each channel and status.

```python
# crosstab is a quick way to count category combinations.
status_crosstab = pd.crosstab(claims_df["channel"], claims_df["status"])
status_crosstab
```

## Your Turn 4: Melt monthly claim counts

Read the wide monthly table. Melt it into `region`, `month`, and `claims`. Clean the month names so they read `jan`, `feb`, and `mar` instead of `jan_claims`, `feb_claims`, and `mar_claims`.

```python
monthly_wide_df = pd.read_csv(monthly_path)

# melt turns month columns into month rows.
monthly_long_df = monthly_wide_df.melt(
    id_vars="region",
    var_name="month",
    value_name="claims",
)
monthly_long_df["month"] = monthly_long_df["month"].str.replace("_claims", "", regex=False)
monthly_long_df
```
