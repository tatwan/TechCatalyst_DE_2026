## Instructor Teaching Note

This activity separates one-number summary questions from grouped summary questions. Ask students which questions need `nunique()`, `min()`, or `max()`, and which questions need `value_counts()`.

```python
import pandas as pd

DATA_PATH = "../Activity_8_Comic_Books_Summary/Resources/books_clean.csv"
books_df = pd.read_csv(DATA_PATH)
books_df.head()
```

## 1. Inspect the cleaned data

Display row count, columns, and dtypes.

```python
print(len(books_df))
print(books_df.columns)
books_df.dtypes
```

## 2. Calculate summary metrics

Calculate unique author count, unique country count, earliest publication year, latest publication year, and total rows.

```python
unique_authors = books_df["Author"].nunique()
unique_countries = books_df["Country of Publication"].nunique()
earliest_year = books_df["Publication Year"].min()
latest_year = books_df["Publication Year"].max()
total_rows = len(books_df)
```

## 3. Build a one-row summary DataFrame

Create `summary_df` with readable column names and one row of values.

```python
# Simple path: one-row DataFrame from a dictionary of lists.
summary_df = pd.DataFrame({
    "unique_authors": [unique_authors],
    "unique_countries": [unique_countries],
    "earliest_year": [earliest_year],
    "latest_year": [latest_year],
    "total_rows": [total_rows],
})
summary_df

# Also valid: a list with one dictionary.
# pd.DataFrame([{
#     "unique_authors": unique_authors,
#     "unique_countries": unique_countries,
#     "earliest_year": earliest_year,
#     "latest_year": latest_year,
#     "total_rows": total_rows,
# }])
```

## 4. Add country counts

Create `country_counts` with the top 10 countries by publication count.

```python
country_counts = books_df["Country of Publication"].value_counts().head(10)
country_counts
```
