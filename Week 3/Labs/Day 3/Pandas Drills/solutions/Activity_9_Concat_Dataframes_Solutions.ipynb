# Activity 9: Concat DataFrames

Combine dues and membership data from two student organizations. This is pandas concat practice and a bridge to SQL `UNION` and joins.

Solutions notebook for instructor reference. Keep the student drill notebook unchanged.

## Instructor Teaching Note

Use this as a bridge to SQL thinking. `pd.concat()` stacks similar rows, like `UNION ALL`; `pd.merge()` matches related rows, like a `JOIN`.

```python
import pandas as pd

RESOURCE_DIR = "../Activity_9_Concat_Dataframes/Resources"
fin_dues_df = pd.read_csv(f"{RESOURCE_DIR}/fin_leaders_america.csv")
inv_dues_df = pd.read_csv(f"{RESOURCE_DIR}/invstrs_leadership.csv")
fin_members_df = pd.read_csv(f"{RESOURCE_DIR}/fin_leaders_members.csv")
inv_members_df = pd.read_csv(f"{RESOURCE_DIR}/invstrs_leadership_members.csv")

fin_dues_df.head()
```

## 1. Inspect all four tables

Display shape and columns for each DataFrame.

```python
print(fin_dues_df.shape, fin_dues_df.columns)
print(inv_dues_df.shape, inv_dues_df.columns)
print(fin_members_df.shape, fin_members_df.columns)
print(inv_members_df.shape, inv_members_df.columns)
```

## 2. Add source labels

Add an `Organization` column to each dues table and each member-status table so rows keep their origin after concat.

```python
fin_dues_df["Organization"] = "Financial Leaders of America"
inv_dues_df["Organization"] = "Investment Leadership"
fin_members_df["Organization"] = "Financial Leaders of America"
inv_members_df["Organization"] = "Investment Leadership"
```

## 3. Concat dues by rows

Stack the two dues tables into `all_dues_df`. Use `ignore_index=True`.

```python
# concat stacks rows when the tables have the same kind of columns.
all_dues_df = pd.concat([fin_dues_df, inv_dues_df], ignore_index=True)
all_dues_df.head()
```

## 4. Concat member status by rows

Stack the two member-status tables into `all_members_df`. Use `ignore_index=True`.

```python
all_members_df = pd.concat([fin_members_df, inv_members_df], ignore_index=True)
all_members_df.head()
```

## 5. Combine dues with member status

Merge `all_dues_df` and `all_members_df` on `MemberName` and `Organization`. Use an inner merge.

```python
# merge matches rows using shared key columns.
master_ledger_df = pd.merge(
    all_dues_df,
    all_members_df,
    on=["MemberName", "Organization"],
    how="inner",
)
master_ledger_df.head()
```

## 6. Create ledger checks

Add `BalanceDue` as `DuesOwed - DuesPaid`. Then summarize total balance by organization.

```python
master_ledger_df["BalanceDue"] = master_ledger_df["DuesOwed"] - master_ledger_df["DuesPaid"]
balance_summary = master_ledger_df.groupby("Organization")["BalanceDue"].sum()
balance_summary
```
