This is a short confidence drill. Filter a utility dataset, sort by usage, and identify the highest-consumption record.

## Instructor Teaching Note

This is a confidence drill. The simple path is filter, sort, take the first row. After that lands, mention that `.idxmax()` is another way to find the highest value.

```python
import pandas as pd

DATA_PATH = "../Activity_6_Search_For_The_Worst/Resources/SFO_Airport_Utility_Consumption.csv"
usage_df = pd.read_csv(DATA_PATH)
usage_df.head()
```

## 1. Inspect categories

Print the unique values in `Utility` and `Owner`.

```python
print(usage_df["Utility"].unique())
print(usage_df["Owner"].unique())
```

## 2. Choose one slice

Create `electricity_tenant_df` with rows where `Utility` is `Electricity` and `Owner` is `Tenant`.

```python
electricity_tenant_df = usage_df[
    (usage_df["Utility"] == "Electricity") & (usage_df["Owner"] == "Tenant")
]
electricity_tenant_df.head()
```

## 3. Sort by usage

Sort from highest usage to lowest usage and reset the index.

```python
worst_usage_df = electricity_tenant_df.sort_values("Usage", ascending=False).reset_index(drop=True)
worst_usage_df.head()
```

## 4. Pull the worst record

Use `.iloc[0]` to save the highest-usage row as `worst_day`.

```python
worst_day = worst_usage_df.iloc[0]
worst_day

# Also valid after students understand sorting:
# worst_day = electricity_tenant_df.loc[electricity_tenant_df["Usage"].idxmax()]
```
