# Activity 4: Crowdfunding Cleanup

Clean and analyze a crowdfunding dataset. This drill combines column selection, filtering, derived columns, type conversion, and groupby.

Solutions notebook for instructor reference. Keep the student drill notebook unchanged.

## Instructor Teaching Note

Keep the flow concrete: select useful columns, filter rows, add one calculated column, then summarize. If students struggle with `groupby().agg()`, have them calculate one metric at a time first.

```python
import pandas as pd

DATA_PATH = "../Activity_4_Crowdfunding_Cleanup/Resources/CrowdfundingData.csv"
crowdfunding_df = pd.read_csv(DATA_PATH)
crowdfunding_df.head()
```

## 1. Keep the working columns

Create `campaigns_df` with these columns: `name`, `goal`, `pledged`, `outcome`, `country`, `staff_pick`, `backers_count`, `spotlight`, and `category`.

```python
campaigns_df = crowdfunding_df[[
    "name", "goal", "pledged", "outcome", "country", "staff_pick",
    "backers_count", "spotlight", "category",
]].copy()
campaigns_df.head()
```

## 2. Filter to campaigns that raised money

Keep rows where `pledged` is greater than 0. Print the before and after row counts.

```python
raised_df = campaigns_df[campaigns_df["pledged"] > 0]

print(len(campaigns_df))
print(len(raised_df))
```

## 3. Focus on US campaigns

Create `us_df` with only campaigns where `country` is `US`.

```python
us_df = raised_df[raised_df["country"] == "US"].copy()
us_df.head()
```

## 4. Add average pledge per backer

Create `average_pledge` as `pledged / backers_count`. Avoid division by zero by using the filtered data from step 2.

```python
us_df["average_pledge"] = us_df["pledged"] / us_df["backers_count"]
us_df[["name", "pledged", "backers_count", "average_pledge"]].head()
```

## 5. Summarize staff picks

Group by `staff_pick` and calculate campaign count, total pledged, average pledge, and success rate. For success rate, count the share where `outcome` equals `successful`.

```python
campaigns_df["is_successful"] = campaigns_df["outcome"] == "successful"

staff_pick_summary = campaigns_df.groupby("staff_pick").agg({
    "name": "count",
    "pledged": ["sum", "mean"],
    "is_successful": "mean",
})
staff_pick_summary
```
