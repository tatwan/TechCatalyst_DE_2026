# Activity 7: Comic Books CSV Cleanup

Clean a large comic books CSV into a smaller, analysis-ready table.

Solutions notebook for instructor reference. Keep the student drill notebook unchanged.

## Instructor Teaching Note

Show one-column cleanup first if needed, then use the loop once students understand the repeated pattern. The teaching goal is not the loop, it is recognizing repeated string cleanup.

```python
from pathlib import Path
import pandas as pd

DATA_PATH = "../Activity_7_Comic_Books_CSV_Cleanup/Resources/comic_books_expanded.csv"
books_raw_df = pd.read_csv(DATA_PATH)
books_raw_df.head()
```

## 1. Inspect the columns

Display the column names and row count.

```python
print(books_raw_df.columns)
print(len(books_raw_df))
```

## 2. Keep the useful columns

Create `books_df` with these columns: `ISBN`, `Title`, `Other titles`, `Name`, `All names`, `Country of publication`, `Place of publication`, `Publisher`, and `Date of publication`.

```python
books_df = books_raw_df[[
    "ISBN", "Title", "Other titles", "Name", "All names",
    "Country of publication", "Place of publication", "Publisher", "Date of publication",
]].copy()
books_df.head()
```

## 3. Rename columns

Rename `Other titles` to `Other Titles`, `Name` to `Author`, and `Date of publication` to `Publication Year`.

```python
books_df = books_df.rename(columns={
    "Other titles": "Other Titles",
    "Name": "Author",
    "All names": "All Names",
    "Country of publication": "Country of Publication",
    "Place of publication": "Place of Publication",
    "Date of publication": "Publication Year",
})
books_df.columns
```

## 4. Clean text fields

Strip leading and trailing whitespace from string columns. Then title-case `Country of Publication` and `Place of Publication`.

```python
for column in books_df.select_dtypes(include="object").columns:
    books_df[column] = books_df[column].str.strip()

books_df["Country of Publication"] = books_df["Country of Publication"].str.title()
books_df["Place of Publication"] = books_df["Place of Publication"].str.title()
books_df.head()
```

## 5. Clean publication year

Convert `Publication Year` to numeric with `errors="coerce"`. Keep rows where publication year is not missing.

```python
before_count = len(books_df)
books_df["Publication Year"] = pd.to_numeric(books_df["Publication Year"], errors="coerce")
books_df = books_df.dropna(subset=["Publication Year"])

print(before_count)
print(len(books_df))
```

## 6. Write a cleaned CSV

Write the cleaned table to `Output/books_clean.csv`. Create the `Output` folder if needed.

```python
output_folder = Path("../Activity_7_Comic_Books_CSV_Cleanup/Output")
output_folder.mkdir(exist_ok=True)

books_df.to_csv(output_folder / "books_clean.csv", index=False)
pd.read_csv(output_folder / "books_clean.csv").head()
```
