```python
import pandas as pd

DATA_PATH = "../Activity_12_Rolling_Window_Signals/Resources/daily_claims_volume.csv"
```

## Example 1: Rolling average

A rolling average smooths nearby rows. It is useful when one high or low day should not dominate the story.

```python
example = pd.DataFrame({"day": [1, 2, 3, 4, 5], "claims": [10, 30, 20, 40, 50]})
example["claims_3day_avg"] = example["claims"].rolling(window=3, min_periods=2).mean()
example
```

## Example 2: Shift for prior-row comparison

`shift(1)` puts the previous row next to the current row, which lets you calculate a day-over-day change.

```python
example["prior_claims"] = example["claims"].shift(1)
example["claim_delta"] = example["claims"] - example["prior_claims"]
example
```

## Your Turn 1: Load and sort the daily data

Read `daily_claims_volume.csv`, parse `date`, and sort by date.

```python
claims_daily_df = pd.read_csv(DATA_PATH, parse_dates=["date"])
claims_daily_df = claims_daily_df.sort_values("date")

claims_daily_df.head()
claims_daily_df.dtypes
```

## Your Turn 2: Add rolling averages

Add `claims_7day_avg` using a 7-day rolling window with `min_periods=3`. Add `amount_7day_avg` for total claim amount too.

```python
claims_daily_df["claims_7day_avg"] = claims_daily_df["claims_filed"].rolling(window=7, min_periods=3).mean()
claims_daily_df["amount_7day_avg"] = claims_daily_df["total_claim_amount"].rolling(window=7, min_periods=3).mean()
claims_daily_df[["date", "claims_filed", "claims_7day_avg"]].head(10)
```

## Your Turn 3: Add prior-day comparison

Add yesterday's claims, a claim-count delta, and percent change.

```python
claims_daily_df["prior_claims"] = claims_daily_df["claims_filed"].shift(1)
claims_daily_df["claims_delta"] = claims_daily_df["claims_filed"] - claims_daily_df["prior_claims"]
claims_daily_df["claims_pct_change"] = claims_daily_df["claims_filed"].pct_change()
claims_daily_df[["date", "claims_filed", "prior_claims", "claims_delta", "claims_pct_change"]].head(10)
```

## Your Turn 4: Create a signal flag

Create `above_trend_flag` where `claims_filed` is at least 10 higher than `claims_7day_avg`. Count how many days are flagged.

```python
claims_daily_df["above_trend_flag"] = claims_daily_df["claims_filed"] >= claims_daily_df["claims_7day_avg"] + 10
print(claims_daily_df["above_trend_flag"].sum())
claims_daily_df[claims_daily_df["above_trend_flag"]]
```
